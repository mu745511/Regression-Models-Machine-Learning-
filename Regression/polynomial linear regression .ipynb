## Importing the libararies 

```python

import pandas as pd

import numpy as np 

import matplotlib.pyplot as plt 

from sklearn.preprocessing import OneHotEncoder , LabelEncoder , StandardScaler

from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split
```


## Importing the datasets


```python

datasets = pd.read_csv("Position_Salaries.csv")

x  = datasets.iloc[ : , 1: -1].values

y = datasets.iloc[ : , -1].values

print(x)
```

## Training the linear regresson model for the whole datasets

```python

lin_reg = LinearRegression()

lin_reg.fit(x,y)


```


## Training the polynomial linear regression for the whole datasets


```python

from sklearn.preprocessing import PolynomialFeatures

poly_reg = PolynomialFeatures( degree = 4 )

poly_x = poly_reg.fit_transform(x)

lin_reg_2 = LinearRegression()

lin_reg_2.fit( poly_x , y )
```


## Visualising the linear regression results


```python

plt.scatter( x , y , color = 'y')

plt.plot( x , lin_reg.predict(x) , color = 'g')

plt.title('results of linear regression (truth or bluff)')

plt.xlabel('level')

plt.ylabel(' salary ')

plt.show()
```


## Visualising the polynomial linear regression results


```python

plt.scatter( x , y , color = 'r')

plt.plot( x , lin_reg_2.predict(poly_x) , color = 'b')

plt.title("result of polynomial linear rgression (truth or bluff)")

plt.xlabel('level')

plt.ylabel('salary')

plt.show()
```

## optional (to get smother curve)

```python

x_grid = np.arange(min(x), max(x), 0.1)

x_grid = x_grid.reshape((len(x_grid), 1))

plt.scatter(x, y, color = 'red')

plt.plot(x_grid, lin_reg_2.predict(poly_reg.fit_transform(x_grid)), color = 'blue')

plt.title('Truth or Bluff (Polynomial Regression)')

plt.xlabel('Position level')

plt.ylabel('Salary')

plt.show()
```
